# Joining  Tables

Here we are going to join the customers and Sales tables

```python
%%sql
SELECT ORDERNUMBER,
SALES,
MONTH_ID,
YEAR_ID,
CUSTOMERNAME,
CITY,
COUNTRY
FROM sales 
JOIN customer ON sales.ORDER_ID=customer.ORDER_ID
LIMIT 10;
```
