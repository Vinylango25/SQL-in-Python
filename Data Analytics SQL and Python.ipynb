## Q6: Succinctly explain the difference between a subquery and a correlated subquery

A subquery is an independent query nested within another SQL statement, returning a result set. It operates standalone to provide data used in SELECT, INSERT, UPDATE, or DELETE operations. In contrast, a correlated subquery references columns from the outer query, executing once per outer row, influencing its result based on the outer query's current row context.

## Q10:   What is the difference between WHERE and HAVING clauses in a query? Provide an example of each use.

In SQL queries, the WHERE and HAVING clauses serve distinct purposes in filtering data based on specified conditions. The WHERE clause is used to filter individual rows from a table before any groupings or aggregations are applied. For instance, in selecting employees from a table where their salary exceeds 50000, the WHERE clause directly applies conditions to individual rows. Conversely, the HAVING clause is utilized after the GROUP BY clause to filter grouped data based on aggregate conditions. For example, when selecting departments with an average salary exceeding 60000, the HAVING clause filters groups created by GROUP BY DepartmentID based on the average salary condition.

## Q 4: Write a function that will take two parameters: a list of integers and a number. The function must return the indices of the two integers in the list that when added together produce the number passed. If none are found return an empty list.

```python
def find_indices_sum(L,num):
    Ind=[]
    for i in range(len(L)):
        for j in range(len(L)):
            if i!=j and i<j and L[i]+L[j]==num:
                Ind.append((i,j))
    return Ind
            
```

## Q 5: Write a function to return the nth number in the Fibonacci sequence. Make sure it can find the 1000th number in a reasonable amount of time.

```python
def fibonacci(n):
    if n <= 0:
        return "Invalid input. Please provide a positive integer."
    elif n == 1:
        return 0  # First Fibonacci number is 0
    elif n == 2:
        return 1  # Second Fibonacci number is 1
    else:
        a, b = 0, 1
        for _ in range(2, n):
            a, b = b, a + b
        return b
```

```python
%%time

# Example usage:
n = 1000
print(f"The {n}th Fibonacci number is: {fibonacci(n)}")
```

## Q 6: Given a list of indeterminate length, sort the list from smallest to largest. Do not use python’s sort function.

```python
def list_sort(arr):
    n = len(arr)
    # Traverse through all elements in the list
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
```

```python
# Example usage:
my_list = [90,64,22,43, 25, 12, 22, 11,7]
print(f"Original list: {my_list}")
list_sort(my_list)
print(f"Sorted list: {my_list}")
```

## Q 8: Describe the differences among the following data types in Python: list, tuple, dict,OrderedDict, and namedtuple.

In Python, lists (list) are mutable sequences that allow for dynamic modification of elements and preserve the order of insertion. Tuples (tuple), on the other hand, are immutable sequences, meaning their elements cannot be changed after creation, but they also maintain the order of elements. Dictionaries (dict) are mutable and unordered collections of key-value pairs, providing efficient lookup and insertion but not preserving the order of elements. OrderedDict from the collections module is a variant of dictionaries that retains the order of insertion, making it useful when order needs to be preserved. Lastly, namedtuple from the collections module creates tuple subclasses with named fields, offering readable and self-documenting code where each field has a fixed position, akin to a lightweight data structure. Understanding these distinctions helps in selecting the appropriate data structure based on requirements such as mutability, order preservation, and ease of access by index or key.

## Q 9: Create a class that implements a working stack data structure.

```python
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            raise IndexError("pop from empty stack")

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None

    def size(self):
        return len(self.items)

# Example usage:
stack = Stack()
print("Is the stack empty?", stack.is_empty())  # Output: True

stack.push(1)
stack.push(2)
stack.push(3)

print("Top element:", stack.peek())  # Output: 3

print("Pop element:", stack.pop())  # Output: 3
print("Pop element:", stack.pop())  # Output: 2

print("Is the stack empty?", stack.is_empty())  # Output: False
print("Size of the stack:", stack.size())  # Output: 1
```

## Q 10: Create a class that implements a working queue data structure.

```python
class Queue:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            raise IndexError("dequeue from empty queue")

    def peek(self):
        if not self.is_empty():
            return self.items[0]
        else:
            return None

    def size(self):
        return len(self.items)

# Example usage:
queue = Queue()
print("Is the queue empty?", queue.is_empty())  # Output: True

queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)

print("Front element:", queue.peek())  # Output: 1

print("Dequeue element:", queue.dequeue())  # Output: 1
print("Dequeue element:", queue.dequeue())  # Output: 2

print("Is the queue empty?", queue.is_empty())  # Output: False
print("Size of the queue:", queue.size())  # Output: 1
```

## BONUS: Create a class that implements a working deque data structure.

```python
class Deque:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def add_front(self, item):
        self.items.insert(0, item)

    def add_rear(self, item):
        self.items.append(item)

    def remove_front(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            raise IndexError("remove_front from empty deque")

    def remove_rear(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            raise IndexError("remove_rear from empty deque")

    def peek_front(self):
        if not self.is_empty():
            return self.items[0]
        else:
            return None

    def peek_rear(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None

    def size(self):
        return len(self.items)

# Example usage:
deque = Deque()
print("Is the deque empty?", deque.is_empty())  # Output: True

deque.add_rear(1)
deque.add_rear(2)
deque.add_front(3)

print("Front element:", deque.peek_front())  # Output: 3
print("Rear element:", deque.peek_rear())    # Output: 2

print("Remove front:", deque.remove_front())  # Output: 3
print("Remove rear:", deque.remove_rear())    # Output: 2

print("Is the deque empty?", deque.is_empty())  # Output: False
print("Size of the deque:", deque.size())      # Output: 1
```
